```python
from langchain_groq import ChatGroq
llm = ChatGroq(
     model="llama-3.3-70b-versatile",  # fast & free
    temperature=0.6
)
```

Language Translation

```python
from langchain_core.prompts import PromptTemplate
template = "Translate the following {language} text \n{text}"
prompt = PromptTemplate(
    input_variables = ['language','text'],
    template = template
)
final_prompt = prompt.format(language="French",
                             text= "How are you?")
reponse = llm.invoke(final_prompt)
print (final_prompt)
print (reponse.content)
```

Topic Explanation

```python
from langchain_core.prompts import PromptTemplate
template = "Explain topic {topic} in simple words"
prompt = PromptTemplate(
    input_variables = ['topic'],
    template = template
)
final_prompt = prompt.format(topic="technology")
response = llm.invoke(final_prompt)
print (response.content)
```

Convert to Bullet points

```python
from langchain_core.prompts import PromptTemplate
template = "Explain the text into bullet points \n {text}"
prompt = PromptTemplate(
    input_variables = ['text'],
    template = template
)
final_prompt = prompt.format(text="Artificial Intelligence is rapidly changing the world. It is being used in healthcare to diagnose diseases, in education to personalize learning, and in businesses to automate repetitive tasks. AI can improve efficiency and accuracy, but it also raises concerns about job loss and ethical issues. Many experts believe that AI will continue to grow and become an essential part of our daily lives in the future.")
response = llm.invoke(final_prompt)
print (response.content)
```
