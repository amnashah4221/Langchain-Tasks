```python
import os
os.environ['GROQ_API_KEY'] = "yourapikey"
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(
     model="llama-3.3-70b-versatile",  # fast & free
    temperature=0.6
)
```

Text Summarizer

```python
user_text = input("Enter text: ")
prompt = f"Summarize this text to 1/3rd:\n{user_text}"
response = llm.invoke(prompt)
print(response.content)
```

AI Question answering Bot

```python
question = input("Ask your question:")
prompt = f"Answer the following question: \n{question}"
response = llm.invoke(prompt)
print(response.content)
```

Grammar Corrector

```python
text = input("Provide text: ")
prompt = f"Correct grammar: \n{text}"
response = llm.invoke(prompt)
print(response.content)
```
