```python
from langchain_groq import ChatGroq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  # fast & free
    temperature=0.6
)
```

Extract basic Information

```python
from langchain_core.prompts import PromptTemplate
template ="""
Extract the name and age from the provided text

Return in JSON format only(no extra text):
{{
  "name": "string",
  "age":"number"
}}

Text:{text}
"""

prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)
```

```python
final_prompt = prompt.format(text = "My name is Amna Shah. I am 21 years old.")
response = llm.invoke(final_prompt)
print(response.content)
```

```python
#clean output
import re
import json
clean_text =re.search(r'\{.*\}', response.content, re.DOTALL).group()
data = json.loads(clean_text)
print(data)
```

```python
print("Name:", data["name"])
```

Extracting Product Details

```python
from langchain_core.prompts import PromptTemplate
import json
import re
template = """
  Extract the following details about the product
  -  Product name
  -  Product price
  -  Product rating

  Return in JSON format (no extra details)
  {{
    "name": "string",
    "price": "number",
    "rating": "float"
  }}

  Text: {text}
"""
prompt = PromptTemplate(
    input_variables=['text'],
    template=template
)

final_prompt = prompt.format(text="ALi bought iPhone 13 in 1200$ and gave 4.5 stars")
response = llm.invoke(final_prompt)
#print(response.content)

clean_text = re.search(r'\{.*\}', response.content, re.DOTALL).group()
data = json.loads(clean_text)

print("Product Name: ",data["name"])
print("Product Price: ",data["price"])
print("Product Rating: ",data["rating"])
```

Extracting Patient Details

```python
from langchain_core.prompts import PromptTemplate
import json
import re

template = """
  Extract the following details from the patient's input:
  - Symptoms (as a list)
  - Duration
  - Severity (overall severity: mild, moderate or severe)

  Return in JSON format (no extra details)
  {{
    "symptoms": [{{
      "symptom": "string",
      "severity": "string"
    }}],
    "duration": "string"
  }}

Rules:
- If severity is not mentioned -> use unknown.
  Text: {text}
"""

prompt = PromptTemplate(input_variables=["text"], template=template)
final_prompt = prompt.format(text="Patient reports fever for 3 days, severe headache, and mild cough.")
response = llm.invoke(final_prompt)

#print(response.content)

clean_text = re.search(r'\{[\s\S]*\}', response.content, re.DOTALL).group()
data = json.loads(clean_text)

print("Symptoms: ",data["symptoms"])
print("Duration: ",data["duration"])

```

E-Commerce Order Extraction (Advanced Task)

```python
from langchain_core.prompts import PromptTemplate
import json
import re

template = """
  Extract the following details from the Order:
  - Customer Name
  - Order Date
  - Product Name
  - Product Price
  - Payment Method
  - Delivered City
  - Delivery Time

  Return in JSON format (no extra details)
  {{
    "customer_name": "string",
    "order_date": "string",
    "items": [{{
      "product_name": "string",
      "product_price": "number"
    }}],
   "payment_method": "string",
   "delivered_city": "string",
   "delivery_time": "string"
  }}

Rules:
- Remove $ from price-> Convert to number
- Must return a list of objects
- Only valid JSON format is allowed
  Text: {text}
"""

prompt = PromptTemplate(input_variables=["text"], template=template)
final_prompt = prompt.format(text="John Doe placed an order on 5th March 2025 for a MacBook Pro worth $1800 and a wireless mouse for $50. He used a credit card for payment and wants delivery in Lahore within 3 days.")
response = llm.invoke(final_prompt)

#print(response.content)

clean_text = re.search(r'\{[\s\S]*\}', response.content, re.DOTALL).group()
data = json.loads(clean_text)

print("Customer Name: ",data["customer_name"])
print("Order Date: ",data["order_date"])
print("Items: ",data["items"])
print("Payment Method:", data["payment_method"])
print("Delivered City: ",data["delivered_city"])
print("Delivery Time: ",data["delivery_time"])

```
