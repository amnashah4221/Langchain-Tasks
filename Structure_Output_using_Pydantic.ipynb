```python
from langchain_groq import ChatGroq
llm = ChatGroq(
     model="llama-3.3-70b-versatile",  # fast & free
    temperature=0.6
)
```

Extracting Order Details

```python
from pydantic import BaseModel
from typing import List

class Item(BaseModel):
  product:str
  price:int
class Order(BaseModel):
  customer_name: str
  order_date: str
  items: List[Item]
  payment_method: str
  delivery_city: str
  delivery_time: str
structured_llm = llm.with_structured_output(Order)
result = structured_llm.invoke(
     "John Doe placed an order on 5th March 2025 for a MacBook Pro worth $1800 and a wireless mouse for $50. He used a credit card for payment and wants delivery in Lahore within 3 days."
)
print(result)

print(result.customer_name)
```

Extracting Flight Details

```python
from pydantic import BaseModel

class Flight(BaseModel):
  person_name: str
  source: str
  destination: str
  date: str
  ticket_price: int
  class_type: str
structured_llm = llm.with_structured_output(Flight)
response = structured_llm.invoke(
    "Ahmed booked a flight from Lahore to Dubai on 12th June. The ticket cost $500 and he chose economy class."
)

print("Name: ",response.person_name)
print("Flight Source: ",response.source)
print("Flght Destination: ",response.destination)
print("Flight Date: ",response.date)
print("Ticket Price: ",response.ticket_price)
print("FLight Type: ",response.class_type)
```

Extracting Restaurant Order Details for Billing

```python
from pydantic import BaseModel
from typing import List
class Item(BaseModel):
  name: str
  quantity: int
  price_per_item: int
class Order(BaseModel):
  items: List[Item]
  total_price: int
structured_llm = llm.with_structured_output(Order)
response = structured_llm.invoke(
    "Fatima ordered 2 pizzas for $10 each and 3 burgers for $5 each."
)

print("\n Order Summary")
print("-" * 30)

for item in response.items:
    print(f"Item: {item.name}")
    print(f"  Quantity: {item.quantity}")
    print(f"  Price per item: ${item.price_per_item}")
    print(f"  Subtotal: ${item.quantity * item.price_per_item}")
    print()

print("-" * 30)
print(f" Total Price: ${response.total_price}")
```
